```python
word.isna().sum()
```

```python
word.dropna(inplace = True)
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
polarity = []
for i in word['comment_text']:
    polarity.append(TextBlob(i).sentiment.polarity)
```

```python
word['polarity'] = polarity
```

```python
word
```

```python
comment_positive = word[word['polarity'] == 1]
```

```python
comment_positive
```

```python
stopwords = set(STOPWORDS)
```

```python
word_cluster =' '.join(comment_positive['comment_text'])
word_cluster
```

```python
wordcloud = WordCloud(width = 1000, height = 500, stopwords = stopwords).generate(word_cluster)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```

```python
comment_neg = word[word['polarity'] == -1]
comment_neg
```

```python
word_neg_cluster = ' '.join(comment_neg['comment_text'])
word_neg_cluster
```

```python
wordcloud_neg = WordCloud(width = 1000, height = 500, stopwords = stopwords).generate(word_neg_cluster)
wordcloud_neg
```

```python
plt.figure(figsize = (15,8))
plt.imshow(wordcloud_neg)
plt.axis('off')
```
